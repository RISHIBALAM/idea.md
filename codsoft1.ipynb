```python
import tkinter as tk
from tkinter import messagebox
import json
import dateparser
import os

TASK_FILE = "tasks.json"

def load_tasks():
    if os.path.exists(TASK_FILE):
        with open(TASK_FILE, 'r') as f:
            return json.load(f)
    return []

def save_tasks(tasks):
    with open(TASK_FILE, 'w') as f:
        json.dump(tasks, f, indent=4)

class TodoApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Smart To-Do List")
        self.tasks = load_tasks()

        self.entry = tk.Entry(root, width=50)
        self.entry.pack(pady=10)

        self.add_button = tk.Button(root, text="Add Task", command=self.add_task)
        self.add_button.pack()

        self.listbox = tk.Listbox(root, selectmode=tk.SINGLE, width=70, height=10)
        self.listbox.pack(pady=10)

        self.done_button = tk.Button(root, text="Mark as Done", command=self.mark_done)
        self.done_button.pack()

        self.delete_button = tk.Button(root, text="Delete Task", command=self.delete_task)
        self.delete_button.pack()

        self.refresh_tasks()

    def add_task(self):
        task_text = self.entry.get()
        if not task_text.strip():
            messagebox.showwarning("Input Error", "Please enter a task.")
            return

        parsed_date = dateparser.parse(task_text)
        task = {
            "task": task_text,
            "done": False,
            "parsed_date": parsed_date.strftime("%Y-%m-%d %H:%M:%S") if parsed_date else None
        }
        self.tasks.append(task)
        save_tasks(self.tasks)
        self.entry.delete(0, tk.END)
        self.refresh_tasks()

    def mark_done(self):
        selected = self.listbox.curselection()
        if not selected:
            messagebox.showwarning("Select Task", "Select a task to mark as done.")
            return
        idx = selected[0]
        self.tasks[idx]["done"] = True
        save_tasks(self.tasks)
        self.refresh_tasks()

    def delete_task(self):
        selected = self.listbox.curselection()
        if not selected:
            messagebox.showwarning("Select Task", "Select a task to delete.")
            return
        idx = selected[0]
        del self.tasks[idx]
        save_tasks(self.tasks)
        self.refresh_tasks()

    def refresh_tasks(self):
        self.listbox.delete(0, tk.END)
        for i, task in enumerate(self.tasks):
            status = "✓" if task["done"] else "✗"
            date_str = f" [Due: {task['parsed_date']}]" if task['parsed_date'] else ""
            self.listbox.insert(tk.END, f"{status} {task['task']}{date_str}")

if __name__ == "__main__":
    root = tk.Tk()
    app = TodoApp(root)
    root.mainloop()
```

```python
import json
import dateparser
import os
from datetime import datetime

TASK_FILE = "tasks.json"

def load_tasks():
    if os.path.exists(TASK_FILE):
        try:
            with open(TASK_FILE, 'r') as f:
                return json.load(f)
        except json.JSONDecodeError:
            return [] # Return empty list if file is empty or invalid JSON
    return []

def save_tasks(tasks):
    with open(TASK_FILE, 'w') as f:
        json.dump(tasks, f, indent=4)

def display_tasks(tasks):
    if not tasks:
        print("No tasks in the list.")
        return

    print("\n--- To-Do List ---")
    for i, task in enumerate(tasks):
        status = "✓" if task["done"] else "✗"
        date_str = f" [Due: {task['parsed_date']}]" if task['parsed_date'] else ""
        print(f"{i+1}. {status} {task['task']}{date_str}")
    print("------------------\n")

def add_task(tasks):
    task_text = input("Enter task: ")
    if not task_text.strip():
        print("Task cannot be empty.")
        return tasks

    parsed_date = dateparser.parse(task_text)
    task = {
        "task": task_text,
        "done": False,
        "parsed_date": parsed_date.strftime("%Y-%m-%d %H:%M:%S") if parsed_date else None
    }
    tasks.append(task)
    save_tasks(tasks)
    print("Task added.")
    return tasks

def mark_done(tasks):
    display_tasks(tasks)
    try:
        task_index = int(input("Enter the number of the task to mark as done: ")) - 1
        if 0 <= task_index < len(tasks):
            tasks[task_index]["done"] = True
            save_tasks(tasks)
            print("Task marked as done.")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a number.")
    return tasks

def delete_task(tasks):
    display_tasks(tasks)
    try:
        task_index = int(input("Enter the number of the task to delete: ")) - 1
        if 0 <= task_index < len(tasks):
            del tasks[task_index]
            save_tasks(tasks)
            print("Task deleted.")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a number.")
    return tasks

def main():
    tasks = load_tasks()

    while True:
        print("Choose an action:")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Mark Task as Done")
        print("4. Delete Task")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            tasks = add_task(tasks)
        elif choice == '2':
            display_tasks(tasks)
        elif choice == '3':
            tasks = mark_done(tasks)
        elif choice == '4':
            tasks = delete_task(tasks)
        elif choice == '5':
            print("Exiting To-Do application.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
